```python
import networkx as nx
import matplotlib.pyplot as plt

# 定义图中的节点和边
# 节点是图中的顶点，边是连接这些顶点的线
users = [1,2,3,4]
edges = [(1,2),(2,3),(3,4),(4,1)]

# 初始化一个无向图
G = nx.Graph()
# 将定义的节点和边添加到图中
G.add_nodes_from(users)
G.add_edges_from(edges)

# 设置图形的尺寸
plt.figure(figsize=(8,8))
# 使用spring_layout算法为图中的节点分配位置
# 这种布局算法会尝试将图形的节点放置在看起来平衡的位置
pos = nx.spring_layout(G)
# with_labels=True表示在节点上显示标签，node_color和edge_color设置节点和边的颜色
# node_size和font_size分别设置节点的大小和字体的大小，font_weight设置字体的粗细
nx.draw(G,pos,with_labels=True,node_color='lightblue',edge_color='gray',node_size=2000,font_size=15,font_weight='bold')
plt.show()
```
